# Necessary Libraries

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

# Assignment 1

```python
# Random Matrix
a_mat = np.random.randint(0,10,size=(3,3))
b_mat = np.random.randint(0,10,size=(3,3))

# Sum of 2 Random Matrix
sum_mat = a_mat + b_mat

# Subtraction of 2 Random Matrix
sub_mat = a_mat - b_mat

# Multiplication of 2 Random Matrix
mul_mat = np.dot(a_mat,b_mat)

# Matrix Diterminent
det_mata = np.linalg.det(a_mat)
det_matb = np.linalg.det(b_mat)

# Matrix Inverse for a
if(np.linalg.det(a_mat)!=0):
  inv_mata = np.linalg.inv(a_mat)
else:
  print("Matrix is Singular")

# Matrix Inverse for b
if(np.linalg.det(b_mat)!=0):
  inv_matb = np.linalg.inv(b_mat)
else:
  print("Matrix is Singular")

# Rank of the Matrix
rank_mata = np.linalg.matrix_rank(a_mat)
rank_matb = np.linalg.matrix_rank(b_mat)

# Print Functions
print("Matrix A: \n",a_mat,"\n")
print("Matrix B: \n",b_mat,"\n")
print("Sum of Matrix: \n",sum_mat,"\n")
print("Subtraction of Matrix: \n",sub_mat,"\n")
print("Multiplication of Matrix: \n",mul_mat,"\n")
print("Matrix Determinant of Mat A: \n",det_mata,"\n")
print("Matrix Determinant of Mat B: \n",det_matb,"\n")
print("Matrix Inverse of Mat A: \n",inv_mata,"\n")
print("Matrix Inverse of Mat B: \n",inv_matb,"\n")
print("Matrix Rank of Mat A: \n",rank_mata,"\n")
print("Matrix Rank of Mat B: \n",rank_matb,"\n")
```

# Assignment 2

```python
# Random Matrix with float
A = np.random.uniform(11,30, size=(5,3))
B = np.random.uniform(31,40, size=(5,3))

print("Matrix A: \n",A,"\n")
print("Matrix B: \n",B,"\n")

# Flattem 1D Array
A_flat = A.flatten()
B_flat = B.flatten()

print("Flattened Matrix A: \n",A_flat,"\n")
print("Flattened Matrix B: \n",B_flat,"\n")

# Create an Index for x-Axis
x = np.arange(len(A_flat))

# Line Plot
plt.plot(x, A_flat, marker = "o", linestyle = "-", label="Matrix A")
plt.plot(x, B_flat, marker = "s", linestyle = "--", label="Matrix B")

# Label and Title
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Line Graph of Float Flatten Matrices A and B")
plt.legend()
plt.grid(True)

# Show the Plot
plt.show()
```
