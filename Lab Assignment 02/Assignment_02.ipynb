# Necessary Libraries

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Creating Personal Matrix as A

```python
# ID Matrix
d1 = 6
d2 = 9

A = np.array([[d1+2, d2+1],[2*d1, d2+2]])

# Show Matrix
print("Matrix A: \n",A,"\n")
```

# 2. Analyze Matrix A

```python
# Shape of A
size_A = A.shape

# Determinant
det_A = np.linalg.det(A)

# Rank
rank_A = np.linalg.matrix_rank(A)

# Eigenvalues
eigenvalA = np.linalg.eigvals(A)

# Inverse (only if determinant ‚â† 0)
if np.linalg.det(A) != 0:
    inv_A = np.linalg.inv(A)
else:
    inv_A = "Matrix is singular, cannot compute inverse"

# Display results
print("Shape of A:", size_A,"\n")
print("Determinant of A:", det_A,"\n")
print("Rank of A:", rank_A,"\n")
print("Eigenvalues of A:", eigenvalA,"\n")
print("Inverse of A:\n", inv_A,"\n")
```

# 3. Value Change Experiment

```python
# New Matrix B
B = np.array([[d1+2, d2+1],[2*d1, d2+3]]) # 4th element as d2+3 instead of d2+2

# Show Matrix
print("Matrix B:\n",B,"\n")
```

# 4. Re-analyze Matrix B

```python
# Shape of B
size_B = B.shape

# Determinant
det_B = np.linalg.det(B)

# Rank
rank_B = np.linalg.matrix_rank(B)

# Eigenvalues
eigenvalB = np.linalg.eigvals(B)

# Inverse (only if determinant ‚â† 0)
if np.linalg.det(B) != 0:
    inv_B = np.linalg.inv(B)
else:
    inv_B = "Matrix is singular, cannot compute inverse"

# Display results
print("Shape of B:", size_B,"\n")
print("Determinant of B:", det_B,"\n")
print("Rank of B:", rank_B,"\n")
print("Eigenvalues of B:", eigenvalB,"\n")
print("Inverse of B:\n", inv_B,"\n")
```

# 5. Comparison & Explanation

**Q1. How did the determinant change and why?**  
A: For a 2√ó2 matrix, determinant = ad ‚àí bc.  
  
      Det(A) = (8√ó11) ‚àí (10√ó12) = 88 ‚àí 120 = ‚àí32  
      Det(B) = (8√ó12) ‚àí (10√ó12) = 96 ‚àí 120 = ‚àí24  

The determinant increased from ‚àí32 to ‚àí24.  
It changed because the determinant depends directly on the values of the matrix elements.When I increased one element (bottom-right) by 1, it increased the product ad, which changed the final subtraction result. Even a small change in one element can significantly affect the determinant.  
  
  
**Q2. Did the rank change? Explain.**  
A: No, the rank did not change.  

Both matrices still have non-zero determinants. For a 2√ó2 matrix, if the determinant is not zero, the rank is 2 (full rank). Since both ‚àí32 and ‚àí24 are non-zero, both matrices remain full rank.  

So the rank stays 2.  
  
  
**Q3. How did the eigenvalues respond to the value change?**  
A: The eigenvalues changed slightly.  

Eigenvalues depend on both the trace (sum of diagonal elements) and the determinant. Since we increased one diagonal element, the trace increased, and the determinant also changed. Because eigenvalues are calculated from these values, they adjusted accordingly.  

So the eigenvalues did not remain the same ‚Äî they shifted due to the modification in the matrix.  


**Q4. Is B easier or harder to invert than A? Why?**  
A: Both matrices are invertible because their determinants are non-zero.  

But, B is slightly harder to invert than A.  

This is because B has a smaller absolute determinant (|‚àí24|) compared to A (|‚àí32|). A smaller determinant means the matrix is closer to singular (zero determinant), which makes the inverse less stable.  
  
So, A is slightly easier (more stable) to invert than B.
